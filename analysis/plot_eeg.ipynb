```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
# from IPython.display import display, HTML
# display(HTML("<style>.container { width:100% !important; } .CodeMirror pre {font-size: 9pt;}</style>"))
```

```python
def smooth(data, step):
    smoothed = []
    for i in range(0, len(data), step):
        # print(i, i+step)
        if i > len(data) - 100:
            smoothed.append(np.mean(data[i:]))
        else:        
            smoothed.append(np.mean(data[i:i+step]))
    return smoothed
```

```python
def plot_eeg(subject):
    file = f'./data/{subject}/attentivu/exg0.csv'
    df = pd.read_csv(file, sep=',')

    # df.plot(kind='line',
    #         x='UTC_TIME', y='CH1',
    #         color='red', figsize=(15,5))
    # plt.show()

    ch1_smooth = smooth(df['CH1'], 1000)
    ch2_smooth = smooth(df['CH2'], 1000)
    ch3_smooth = smooth(df['CH3'], 1000)

    fig, ax = plt.subplots(1,3)#,figsize=(60, nrecs*8))

    ax[0,0].set_title(subject)
    ax[0,0].plot(ch1_smooth)
    ax[0,1].plot(ch2_smooth)
    ax[0,2].plot(ch3_smooth)

    # plt.ylim([-10000,10000])
    # plt.show()
    # plt.plot(ch2_smooth)
    # plt.ylim([-10000,10000])
    # plt.show()
    # plt.plot(ch3_smooth)
    # plt.ylim([-10000,10000])
    # plt.show()
```

```python

subjects = ['claire', 'daniel', 'david', 'khiem', 'michelle', 'nader', 'natasha', 'sven', 'zoey', 'laila'] #'sam', 'harry'

# f = open(f"/Users/karimeladl/Desktop/{subject}/blueberry/0timing.pkl", "rb")
# d = pickle.load(f)

# df = pd.read_csv(f"~/Desktop/Streamer 2/data/{subject}/exg0.csv", "rb")
# df = pd.read_csv(f'../Streamer/data/test_conn/exg0.csv', sep=',')

for s in subjects:
   plot_eeg(s)
```
